```python
##Plot des fonctions chapeau pour un maillage de 5 noeuds
Maillage = np.linspace(0,1,5)
def chapeau0(X):
    res = []
    for x in X:
        if x>=0 and x<=0.25:
            res.append((0.25 -x)/0.25)
        else:
            res.append(0)
    return np.array(res)

def chapeau1(X):
    res = []
    for x in X:
        if x>=0 and x<=0.25:
            res.append((x)/0.25)
        elif x>0.25 and x<=0.5:
            res.append((0.5-x)/0.25)
        else:
            res.append(0)
    return np.array(res)

def chapeau2(X):
    res = []
    for x in X:
        if x>=0.25 and x<=0.5:
            res.append((x-0.25)/0.25)
        elif x>0.5 and x<=0.75:
            res.append((0.75-x)/0.25)
        else:
            res.append(0)
    return np.array(res)

def chapeau3(X):
    res = []
    for x in X:
        if x>=0.5 and x<=0.75:
            res.append((x-0.5)/0.25)
        elif x>0.75 and x<=1:
            res.append((1-x)/0.25)
        else:
            res.append(0)
    return np.array(res)

def chapeau4(X):
    res = []
    for x in X:
        if x>=0.75 and x<=1:
            res.append((x-0.75)/0.25)
        else:
            res.append(0)
    return np.array(res)
    
plt.figure(figsize=(12,4))
plt.plot(Maillage, chapeau0(Maillage), label="phi0")
plt.plot(Maillage, chapeau1(Maillage), label="phi1")
plt.plot(Maillage, chapeau2(Maillage), label="phi2")
plt.plot(Maillage, chapeau3(Maillage), label="phi3")
plt.plot(Maillage, chapeau4(Maillage), label="phi4")
plt.title("Fonctions chapeau pour un maillage de 5 points")
plt.legend()
plt.savefig("chapeau.png")
plt.show()
```

```python
##Exemple des fonctions fonctions de base locales
def phi_local1(X, maillage):
    res = []
    if maillage==0:
        for x in X:
            if x>=0 and x<=0.25:
                res.append((0.25 -x)/0.25)
        return res
    if maillage==1:
        for x in X:
            if x>=0.25 and x<=0.5:
                res.append((0.5 -x)/0.25)
        return res
    if maillage==2:
        for x in X:
            if x>=0.5 and x<=0.75:
                res.append((0.75 -x)/0.25)
        return res
    if maillage==3:
        for x in X:
            if x>=0.75 and x<=1:
                res.append((1 -x)/0.25)
        return res
        
def phi_local2(X, maillage):
    res = []
    if maillage==0:
        for x in X:
            if x>=0 and x<=0.25:
                res.append((x)/0.25)
        return res
    if maillage==1:
        for x in X:
            if x>=0.25 and x<=0.5:
                res.append((x-0.25)/0.25)
        return res
    if maillage==2:
        for x in X:
            if x>=0.5 and x<=0.75:
                res.append((x-0.5)/0.25)
        return res
    if maillage==3:
        for x in X:
            if x>=0.75 and x<=1:
                res.append((x-0.75)/0.25)
        return res
color = ["red", "orange", "yellow", "green"]
plt.figure(figsize=(12,4))
for i, col in enumerate(color):
    plt.plot(Maillage[i:i+2], phi_local1(Maillage[i:i+2], i), color=col)
    plt.plot(Maillage[i:i+2], phi_local2(Maillage[i:i+2], i), color=col)
plt.title("Répresentation des fonctions de base locales pour chacun des maillages")
plt.savefig("philoc.png")
plt.show()
```

```python
##Implementation MEF
def u_ex(X):
    return X**2 - X 
def f(X):
    '''
    Fonction source
    '''
    return X**2 - X -2 

def L_i_5(X, i):
    if i == 0:
        return f(X)*chapeau0(X)
    if i == 1:
        return f(X)*chapeau1(X)
    if i == 2:
        return f(X)*chapeau2(X)
    if i == 3:
        return f(X)*chapeau3(X)
    if i == 4:
        return f(X)*chapeau4(X)
        
def L_i_11(X, i):
    if i == 0:
        return f(X)*chapeau0_(X)
    if i == 1:
        return f(X)*chapeau1_(X)
    if i == 2:
        return f(X)*chapeau2_(X)
    if i == 3:
        return f(X)*chapeau3_(X)
    if i == 4:
        return f(X)*chapeau4_(X)
    if i == 5:
        return f(X)*chapeau5_(X)
    if i == 6:
        return f(X)*chapeau6_(X)
    if i == 7:
        return f(X)*chapeau7_(X)
    if i == 8:
        return f(X)*chapeau8_(X)
    if i == 9:
        return f(X)*chapeau9_(X)
    if i == 10:
        return f(X)*chapeau10_(X)

def L_i_21(X, i):
    if i == 0:
        return f(X)*chapeau0__(X)
    if i == 1:
        return f(X)*chapeau1__(X)
    if i == 2:
        return f(X)*chapeau2__(X)
    if i == 3:
        return f(X)*chapeau3__(X)
    if i == 4:
        return f(X)*chapeau4__(X)
    if i == 5:
        return f(X)*chapeau5__(X)
    if i == 6:
        return f(X)*chapeau6__(X)
    if i == 7:
        return f(X)*chapeau7__(X)
    if i == 8:
        return f(X)*chapeau8__(X)
    if i == 9:
        return f(X)*chapeau9__(X)
    if i == 10:
        return f(X)*chapeau10__(X)
    if i == 11:
        return f(X)*chapeau11__(X)
    if i == 12:
        return f(X)*chapeau12__(X)
    if i == 13:
        return f(X)*chapeau13__(X)
    if i == 14:
        return f(X)*chapeau14__(X)
    if i == 15:
        return f(X)*chapeau15__(X)
    if i == 16:
        return f(X)*chapeau16__(X)
    if i == 17:
        return f(X)*chapeau17__(X)
    if i == 18:
        return f(X)*chapeau18__(X)
    if i == 19:
        return f(X)*chapeau19__(X)
    if i == 20:
        return f(X)*chapeau20__(X)
        
for N in [5,11, 21]: #nombre de noeuds à générer
    h = 1/N #definition de la taille du maillage
    a = 1.0 #définition de la consante a du problème
    if N == 5:
        noeud_5 = np.linspace(0,1, N) #Coordonnées du noeud
    elif N == 11:
        noeud_11 = np.linspace(0,1,N)
    elif N == 21:
        noeud_21 = np.linspace(0,1,N)
        
    element = np.empty([N-1, 2]) #Création matrice contenant chque maillage 
    element[:,0], element[:,1] = list(range(1, N)), list(range(2, N+1)) #Assignation selon coordonnées globales
    
    #Création des matrices locales, comme on utilise un maillage uniforme on va utiliser la
    #la même matrice pour tous les maillages de façon practique.
    Ah = 1/h*np.array([[1., -1.],[-1., 1.]])
    Bh = (a*h)/3*np.array([[1., 0.5],[0.5,1.]])
    
    A, B = np.empty([N,N]), np.empty([N,N]) #Déclaration des matrices globales
    for i in np.unique(element):
        look = 0
        for j in element.flatten():
            if i == j:
                look += 1
        if (look == 1 and i == 1):
            for k in range(N):
                if k == 0:
                    A[0, k] = Ah[0, 0]
                    B[0, k] = Bh[0, 0]
                elif k == 1:
                    A[0, k] = Ah[0, 1]
                    B[0, k] = Bh[0, 1]
                else:
                    A[0, k] = 0
                    B[0, k] = 0
        elif(look == 1 and i == N):
            for k in range(N):
                if k == N-1:
                    A[N-1, k] = Ah[1, 1]
                    B[N-1, k] = Bh[1, 1]
                elif k == N-2:
                    A[N-1, k] = Ah[1, 0]
                    B[N-1, k] = Bh[1, 0]
                else:
                    A[N-1, k] = 0
                    B[N-1, k] = 0
        elif look == 2:
            for k in range(1, N-1):
                for l in range(N):
                    if l == k:
                        A[k, l] = Ah[0,0] + Ah[1,1]
                        B[k,l] = Bh[0,0] + Bh[1,1]
                    elif l == k-1:
                        A[k, l] = Ah[0,1]
                        B[k,l] = Bh[0,1]
                    elif l == k+1:
                        A[k,l] = Ah[1,0]
                        B[k,l] = Bh[1,0]
                    elif l < k-1:
                        A[k, l] = 0
                        B[k,l] = 0
                    elif l > k+1:
                        A[k, l] = 0
                        B[k,l] = 0

    X_petit = np.linspace(0,1, 100) 
    L = np.empty([N])
    if N == 5:
        for i in range(N):
            if i ==0:
                L[i] = integrate.trapz(L_i_5(X_petit, i), X_petit) +1
            elif i == N-1:
                L[i] = integrate.trapz(L_i_5(X_petit, i), X_petit) + 1
            else:
                L[i] = integrate.trapz(L_i_5(X_petit, i), X_petit)
        U_5 = np.linalg.solve(A+B, L)
    elif N == 11:
        for i in range(N):
            if i ==0:
                L[i] = integrate.trapz(L_i_11(X_petit, i), X_petit) + 1
            elif i == N-1:
                L[i] = integrate.trapz(L_i_11(X_petit, i), X_petit) + 1
            else:
                L[i] = integrate.trapz(L_i_11(X_petit, i), X_petit)
        U_11 = np.linalg.solve(A+B,L)
    elif N == 21:
        for i in range(N):
            if i ==0:
                L[i] = integrate.trapz(L_i_21(X_petit, i), X_petit) +1
            elif i == N-1:
                L[i] = integrate.trapz(L_i_21(X_petit, i), X_petit) + 1
            else:
                L[i] = integrate.trapz(L_i_21(X_petit, i), X_petit)
        U_21 = np.linalg.solve(A+B,L) #Cer morceaux de code me pose des problèmes, j'obtiens une fonction trés similaire à l'originelle
        #mais déplacé en haut à une constante prés. Comme j'ai implementé la méthode d'une façon très rudimentaire "hard-coding" je ne sais pas si le
        #problème est à cause des fonctions de base globales, la matrice A ou simplement un erreur dans la fonction integrate.trapz.

plt.figure(figsize=(12,4))
plt.plot(X_petit, u_ex(X_petit), "--", label="Solution exacte")
plt.plot(noeud_5, U_5, "-s",label="Solution approchée par 5 noeuds")
plt.plot(noeud_11, U_11, "-o", label="Solution approchée par 11 noeuds")
#plt.plot(noeud_21, U_21, "-d", label="Solution approchée par 21 noeuds")
plt.legend()
plt.title("Comparaison entre les graphes de la solution exacte et les solutions approchées")
plt.savefig("MEF1D.png")
plt.show()
```
